```python
# прочитаем файл в массив
x = []
with open('onegin.txt', 'r') as f:
    x = f.readlines()
```

```python
#получим номера строк с номерами частей стиха (с римскими цифрами)
import re

s1 = r'[XIVL]+'
s2 = r'[^XIVL\s]+'
index_part = []
for i in range(len(x)):
    if (len(re.findall(s1,x[i]))>0) & (len(re.findall(s2,x[i]))==0):
        #print(x[i])
        index_part.append(i)    

```

```python
#поделим массив на части и отберем строки со стихами и  (отберем по наличию двух табуляций в начале) 
data_glava = []
for i in range(len(index_part)-1):
    data_glava.append(list(map(lambda x: x[2:], list(filter(lambda y: y[:2] =='\t\t', x[index_part[i]:index_part[i+1]]))))) 
```

```python
len(data_glava)
```

```python
#
data_glava_l = [[j.lower() for j in i] for i in data_glava]
```

```python
#
data_glava_ll = list(map(lambda y: list(map(lambda x: re.sub(r'[^абвгдеёжзийклмнопрстуфхцчшщъыьэюя\n ]','',x), y)), data_glava_l))
    
```

```python
#
data_glava_lll = list(map(lambda x: list(map(lambda w: re.sub(r'  +',' ',w.strip()), list(filter(lambda y: y!='\n' ,x)))),data_glava_ll))
```

```python
data_trigram = []
for s in data_glava_llll:
    res = []
    #print(len(s.split(' ')))
    for w in s.split(' '):
        i = 2
        if len(w)==1:
            res.append(' '+w+' ')
        elif len(w)==2:
            res.append(' '+w)
        else:     
            res.append(' '+w[:i])          
            while i+2 < len(w):
                res.append(w[i:i+3]) 
                i+=3
            if len(w) - i == 2:
                res.append(w[i:]+' ')
            elif len(w) - i == 1:
                res.append(w[i:]+'  ')
    #print(len(res))               
    data_trigram.append(res)
```

```python
data_bigram = []
for s in data_glava_llll:
    res = []
    #print(len(s.split(' ')))
    for w in s.split(' '):
        i = 1
        if len(w)==1:
            res.append(' '+w)
        else:     
            res.append(' '+w[:i])          
            while i+1 < len(w):
                res.append(w[i:i+2]) 
                i+=2
            if len(w) - i == 1:
                res.append(w[i:]+' ')
    #print(len(res))               
    data_bigram.append(res)
```

```python
data_unogram = []
for i in data_glava_lll:
    data_unogram.append([c for c in ' '.join(i[:5])])
    data_unogram.append([c for c in ' '.join(i[5:9])])
    data_unogram.append([c for c in ' '.join(i[9:])])
```

```python
import torch
```

```python
MAX_LEN_TRI = 150
MAX_LEN_BI = 203
MAX_LEN = 140
```

```python
def data_encode(text, MAX_LEN, CHAR_TO_INDEX):
    X = torch.zeros((len(text), MAX_LEN), dtype=int)
    for i in range(len(text)):
        for j, w in enumerate(text[i]):
            if j >= MAX_LEN:
                break
            X[i, j] = CHAR_TO_INDEX.get(w, CHAR_TO_INDEX['_'])
    return X       
```

```python
X_TRI = data_encode(data_trigram, MAX_LEN_TRI,CHAR_TO_INDEX_TRI)
X_BI = data_encode(data_bigram, MAX_LEN_BI,CHAR_TO_INDEX_BI)
X = data_encode(data_unogram, MAX_LEN,CHAR_TO_INDEX)
```

```python
class Network(torch.nn.Module):

    def __init__(self, len_emb,len_INDEX_TO_CHAR):
        super(Network, self).__init__()
        self.word_embeddings = torch.nn.Embedding(len_INDEX_TO_CHAR, len_emb)
        self.gru = torch.nn.LSTM(len_emb, 256, batch_first=True)
        self.hidden2tag = torch.nn.Linear(256, len_INDEX_TO_CHAR)

    def forward(self, sentences):
        embeds = self.word_embeddings(sentences)
        gru_out, state = self.gru(embeds)
        tag_space = self.hidden2tag(gru_out.reshape(-1, 256))
        return tag_space.reshape(sentences.shape[0], sentences.shape[1], -1), state

    def forward_state(self, sentences, state):
        embeds = self.word_embeddings(sentences)
        gru_out, state = self.gru(embeds, state)
        tag_space = self.hidden2tag(gru_out.reshape(-1, 256))
        return tag_space.reshape(sentences.shape[0], sentences.shape[1], -1), state
```

```python
model = Network(250, len(INDEX_TO_CHAR_BI))
```

```python
import random
def generate_sentence(model,X, INDEX_TO_CHAR, MAX_LEN, CHAR_TO_INDEX):
    j = random.randint(1,len(INDEX_TO_CHAR)-1)
    sentence = [INDEX_TO_CHAR[j]]
    state = None
    for i in range(MAX_LEN):
        X = torch.Tensor([[CHAR_TO_INDEX[sentence[i]]]]).type(torch.long)
        if i == 0:
            result, state = model.forward(X)
        else:
            result, state = model.forward_state(X, state)
        prediction = result[0, -1, :]
        index_of_prediction = prediction.argmax()
        if i >= len(sentence) - 1:
            if index_of_prediction == 0:
                break
        sentence.append(INDEX_TO_CHAR[index_of_prediction])
    print(''.join(sentence))
```

```python
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=.05)
```

```python
len(X)
```

```python
import time

def train(model,X, criterion, optimizer, INDEX_TO_CHAR,MAX_LEN,CHAR_TO_INDEX, epoch):
    for ep in range(1000):
        start = time.time()
        train_loss = 0.0
        train_passed = 0

        for i in range(int(len(X) / 25)):
            batch = X[i * 25:(i + 1) * 25]
            X_batch = batch[:, :-1]
            Y_batch = batch[:, 1:].flatten()

            optimizer.zero_grad()
            answers, _ = model.forward(X_batch)
            answers = answers.view(-1, len(INDEX_TO_CHAR))
            loss = criterion(answers, Y_batch)
            train_loss += loss.item()

            loss.backward()
            optimizer.step()
            train_passed += 1

        print("Epoch {}. Time: {:.3f}, Train loss: {:.3f}".format(ep, time.time() - start, train_loss / train_passed))
        generate_sentence(model,X, INDEX_TO_CHAR, MAX_LEN, CHAR_TO_INDEX)
```

```python

epoch = 1000

train(model,X_BI, criterion,optimizer, INDEX_TO_CHAR_BI, MAX_LEN_BI, CHAR_TO_INDEX_BI, epoch)
```
